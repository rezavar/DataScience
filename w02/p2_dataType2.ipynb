حذف یک عضو از set با استفاده از discard انجام میوشد

```python
set1 = {'reza','ali','saeid','saeid'}
for name in set1:
    print(name)
set1.discard("saeid")  
print(set1)
```

```python
set1 = {1,2,3,4}
set2={4,5,6,7}
set3 = set1|set2
print(set3)
set3 = set1-set2
print(set3)
set3 = set1^set2
print(set3)
```
