رسم نمودار های متنوع matplotlib
مثال و داکیومنت https://matplotlib.org/

plt.plot(x,y,'b')
x,y میتواند لیست یا آرایه باشد
'b' رنگ نمودار 


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5 , 6)
y = x ** 2
plt.plot(x, y,'b')
plt.xlabel(' X ')
plt.ylabel(' X^2 ')
plt.title('pow')
plt.show()
```

میتوان نمودار را به عنوان زیر مجموعه جدا مشخص کرد
plt.subplots()

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)

fig, ax = plt.subplots()
print(fig)
ax.plot(x, y,'r')
plt.xlabel('X ')
plt.ylabel('Y ')
plt.title('title')
plt.show()
```

تعریف ساب پلات چندتایی 
subplots(2,2) دو ردیف و دو ستون
ایجاد فاصله بین نمودارها fig.tight_layout(pad=3)
نمایش نمودار به صورت خط چین قرمز r--

```python
x = [0,1,2,3,4,6,7,8,9,10]
y1 = x
y2 = [i*2 for i in x]
y3 = [i**2 for i in x]
y4 = [i**(1/2) for i in x]

fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(2,2)
fig.tight_layout(pad=3)
ax1.plot(x, y1,'r*')
ax1.set_title('1')

ax2.plot(x, y2,'r--')
ax2.set_title('2')

ax3.plot(x, y3,'b')
ax3.set_title('3')


ax4.plot(x, y4,'b*-')
ax4.set_title('4')

plt.show()
```
