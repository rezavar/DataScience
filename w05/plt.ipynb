```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)

fig, ax = plt.subplots()
print(fig)
ax.plot(x, y,'r')
plt.xlabel('X ')
plt.ylabel('Y ')
plt.title('title')
plt.show()
```

تعریف ساب پلات چندتایی 
subplots(2,2) دو ردیف و دو ستون
ایجاد فاصله بین نمودارها fig.tight_layout(pad=3)
نمایش نمودار به صورت خط چین قرمز r--

```python
x = [0,1,2,3,4,6,7,8,9,10]
y1 = x
y2 = [i*2 for i in x]
y3 = [i**2 for i in x]
y4 = [i**(1/2) for i in x]

fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(2,2)
fig.tight_layout(pad=3)
ax1.plot(x, y1,'r*')
ax1.set_title('1')

ax2.plot(x, y2,'r--')
ax2.set_title('2')

ax3.plot(x, y3,'b')
ax3.set_title('3')


ax4.plot(x, y4,'b*-')
ax4.set_title('4')

plt.show()
```

محل ترسیم نمودار fig.add_axes([0, 0, 1, 1])


```python
x = np.linspace(0, 5, 11)
y = x ** 2
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
plt.show()
```

اضافه کردن راهنما نمودار
اضافه کردن دو خط نمودار در یک فیگور

```python
fig = plt.figure()
x = np.linspace(0, 5, 10)
ax = fig.add_axes([0,0,2,1])
ax.plot(x, x**2, label="x^2")
ax.plot(x, x**3, label="x^3")
ax.legend()
# ax.legend(loc=0) # let matplotlib decide the optimal location
# ax.legend(loc=1) # upper right corner
# ax.legend(loc=2) # upper left corner
# ax.legend(loc=3) # lower left corner
plt.show()
```

رنگ در نمودار ها

```python
x = np.linspace(0, 5, 6)
fig, ax = plt.subplots()

ax.plot(x, x+1, color="b", alpha=0.25) # transparent 25%
ax.plot(x, x+2, color="b", alpha=0.5) # transparent 50%
ax.plot(x, x+3, color="blue", alpha=0.75) # transparent 75%
ax.plot(x, x+4, color="#8B008B")        # RGB hex code
ax.plot(x, x+5, color="#FF8C00")        # RGB hex code 
```
