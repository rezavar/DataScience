متد arange 
یک لیست درست میکند از شماره شروع تا انتها با استپ مشخص شده


```python
print(np.arange(1,10))
print(np.arange(1,10,2))
```

درست کردن و پرکردن یک ارایه با صفر یا یک
سطر و ستون آرایه باید داخل یک پرانتز قرار بگیرد

```python
print(np.zeros(3),"\n\n")
print(np.zeros((3,3)),"\n\n")


print(np.ones(2),"\n\n")
print(np.ones((2,3)),"\n\n")
```

متد linspace
مقادیر ورودی : شروع ، پایان ، تعداد نقطه
ین تابع یک نقطه شروع و پایان میگیرد و تعداد نقاط این تعداد نقاط رو با فاصله مساوی بین شروع و پایان قرار میدهد . 
نکته : خود نقطه شروه و پایان جزو تعداد نقاط فاصله حساب میشوند
li = np.linspace(0,10,3)
بین 0 تا 10 را به دو قسمت مساوی تقسیم میکند

```python
li = np.linspace(0,10,6)
display(li)
```

درست کردن و پرکردن یک آرایه با مقدار رندم بین 0 تا 1
 درست کردن و پرکردن یک ماتریس با مقدار رندم بین 0 تا 1

-------------

یک عدد رندم بین دو عدد
یک آرایه رندم با طول مشخص بین دو عدد
یک ماتریس با سطر و ستون مشخص پر شده با عداد رندم بین دو عدد


```python
display(np.random.rand(2))
display(np.random.rand(3,2))

display(np.random.randint(1,20))
display(np.random.randint(1,20,10))
display(np.random.randint(1,20,(2,3)))
```

تعریف یک آرایه یک بعدی و تبدیل آن به یک ماتریس

```python
arr1 = np.arange(1,25)
arr1 = arr1.reshape(4,6)
arr1
```

یک عدد رندم بین 0 تا 100

ده عدد رندم بین 20 تا 50

```python
arr2 = np.random.randint(20,50,10)
display(arr2)
print("max Number is: {} in position:{}".format(arr2.max(),arr2.argmax()))
print(f"min Number is: {arr2.min()} in position:{arr2.argmin()}")
arr2[0]
```
