```python
str = "one | two"
ls = str.split('|')
print(ls)
```

```python
str = "2 3 1 6 5 4"
ls = str.split()
ls.sort()
print(ls)

ls.sort(reverse=True)
print(ls)
```

```python
cars = [
  {'car': 'Ford', 'year': 2005},
  {'car': 'Mitsubishi', 'year': 2000},
  {'car': 'BMW', 'year': 2019},
  {'car': 'VW', 'year': 2011}
]

def myFunc(e):
  return e['year']

cars.sort(key=myFunc)
print(cars)
```

```python
cars = [
  {'car': 'Ford', 'year': 2005},
  {'car': 'Mitsubishi', 'year': 2000},
  {'car': 'BMW', 'year': 2019},
  {'car': 'VW', 'year': 2011}
]

cars.sort(key=lambda item:item['year'])
print(cars)
```
